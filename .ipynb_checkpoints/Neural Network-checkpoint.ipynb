```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from math import e, floor
```

```python
class Neural_Network:
    def __init__(self, max_iteration=500):
        self.input_layer_node=2
        self.output_layer_node=1
        self.hidden_layer_node=5
        self.learning_rate=0.01
        self.max_iteration=max_iteration
        
        #initialize weights
        self.W1 = np.random.randn(self.input_layer_node,self.hidden_layer_node)
        self.W2 = np.random.randn(self.hidden_layer_node,self.hidden_layer_node)
        self.W3 = np.random.randn(self.hidden_layer_node,self.output_layer_node)
    
    def forward(self,X):    
        self.z2 = np.dot(X,self.W1)
        self.a2 = self.sigmoid(self.z2)
        
        self.z3 = np.dot(self.a2,self.W2)
        self.a3 = self.sigmoid(self.z3)
        
        self.z4 = np.dot(self.a3,self.W3)
        self.yHat = self.sigmoid(self.z4)
        return self.yHat
    
    def predict(self,X):
        self.z2 = np.dot(X,self.W1)
        self.a2 = self.sigmoid(self.z2)
        
        self.z3 = np.dot(self.a2,self.W2)
        self.a3 = self.sigmoid(self.z3)
        
        self.z4 = np.dot(self.a3,self.W3)
        self.yHat = self.sigmoid(self.z4)
        yThresh = self.yHat
        yThresh[yThresh >= 0.5] = 1
        yThresh[yThresh < 0.5] = 0
        return self.yHat

    def predict_proba(self, X):
        return self.forward(X)[-1].T

    def sigmoid(self,z):
        return 1 / (1+e**(-z))
        
    def sigmoid_derivative(self,z):
        sig = self.sigmoid(z)
        return sig*((1-sig))
 
    def backpropgation(self,X,y):
        self.yHat = self.forward(X)
        delta4 = np.multiply((y-self.yHat), self.sigmoid_derivative(self.z4))
        djdW3 = np.dot(self.a3.T,delta4)
        self.W3 = self.W3+djdW3*self.learning_rate
        
        delta3 = np.multiply(np.dot(delta4,self.W3.T),self.sigmoid_derivative(self.z3))
        djdW2 = np.dot(self.a2.T,delta3)
        #print(djdW2.shape)
        self.W2 += djdW2*self.learning_rate

        delta2 = np.multiply(np.dot(delta3,self.W2.T),self.sigmoid_derivative(self.z2))
        djdW1 = np.dot(X.T,delta2)
        self.W1 += djdW1*self.learning_rate

    def errorFunction(self,X,y):
        self.yHat = self.forward(X)
        E = 0.5*sum(np.square(y-self.yHat))
        return E

    def costFunctionPrime(self,X,y):
        self.yHat = self.forward(X)
        delta = np.multiply((y-self.yHat),self.sigmoid_derivative(self.z1))
        djdW1 = np.dot(np.transpose(X),delta)
        return djdW1
    
    def fit(self,X,y):
        costHist = []
        for i in range(self.max_iteration):
            self.backpropgation(X,y)
            costHist.append(self.errorFunction(X,y))
            pass
        pass
```
